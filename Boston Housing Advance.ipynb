```python
df['CRIM'].fillna((df['CRIM'].mean()), inplace=True)
df['ZN'].fillna((df['ZN'].mean()), inplace=True)
df['INDUS'].fillna((df['INDUS'].mean()), inplace=True)
df['CHAS'].fillna((df['CHAS'].median()), inplace=True)
df['AGE'].fillna((df['AGE'].mean()), inplace=True)
df['LSTAT'].fillna((df['LSTAT'].mean()), inplace=True)
```

## Getting X and Y

```python
x=df1.drop("MEDV", axis=1)
y=df1['MEDV'].copy()
```

## Pipeline and Simple Imputer

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline=Pipeline([('Imputer',SimpleImputer()),('std_scalar',StandardScaler())])
```

```python
df_num=my_pipeline.fit_transform(x)
```

```python
df_num.shape
```

## Using of Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model= DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(df_num,y)
```

```python
some_x=x.iloc[:5]
some_y=y.iloc[:5]
prepared_data=my_pipeline.transform(some_x)
model.predict(prepared_data)
```

```python
list(some_y)
```

```python
from sklearn.metrics import mean_squared_error
x_pred=model.predict(df_num)
lin_mse=mean_squared_error(y,x_pred)
lin_rmse=np.sqrt(lin_mse)
lin_rmse
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model,df_num,y,scoring="neg_mean_squared_error",cv=10)
rmse=np.sqrt(-score)
rmse
```

```python
def print_S(score):
    print("score:",score)
    print("Mean",score.mean())
    print("std:",score.std())
print_S(rmse)    
```

## Testing the Model

```python
x_test=strat_test_set.drop("MEDV",axis=1)
y_test=strat_test_set["MEDV"].copy()
x_prepared=my_pipeline.transform(x_test)
final_pred=model.predict(x_prepared)
final_mse=mean_squared_error(y_test,final_pred)
final_rsme=np.sqrt(final_mse)
final_rsme
```
